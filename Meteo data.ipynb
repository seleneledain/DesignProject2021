## Seasonly

```python
# Monthly

monthly_meteo_df = meteo_df.drop(["h"], axis = 1)

# winter season : december, january, february (12,1,2)

idx_win = meteo_df[meteo_df["m"] == 12].index
idx_win2 = meteo_df[meteo_df["m"] == 1].index
idx_win3 = meteo_df[meteo_df["m"] == 2].index

monthly_meteo_df.loc[idx_win, "m"] = 1
monthly_meteo_df.loc[idx_win2, "m"] = 1
monthly_meteo_df.loc[idx_win3, "m"] = 1


# spring season : march, april, may
idx_spr = meteo_df[meteo_df["m"] == 3].index
idx_spr2 = meteo_df[meteo_df["m"] == 4].index
idx_spr3 = meteo_df[meteo_df["m"] == 5].index

monthly_meteo_df.loc[idx_spr, "m"] = 2
monthly_meteo_df.loc[idx_spr2, "m"] = 2
monthly_meteo_df.loc[idx_spr3, "m"] = 2

# summer season : june, july, august
idx_sum = meteo_df[meteo_df["m"] == 6].index
idx_sum2 = meteo_df[meteo_df["m"] == 7].index
idx_sum3 = meteo_df[meteo_df["m"] == 8].index

monthly_meteo_df.loc[idx_sum, "m"] = 3
monthly_meteo_df.loc[idx_sum2, "m"] = 3
monthly_meteo_df.loc[idx_sum3, "m"] = 3

# FAAALL season : september, october, november
idx_aut = meteo_df[meteo_df["m"] == 9].index
idx_aut2 = meteo_df[meteo_df["m"] == 10].index
idx_aut3 = meteo_df[meteo_df["m"] == 11].index

monthly_meteo_df.loc[idx_aut, "m"] = 4
monthly_meteo_df.loc[idx_aut2, "m"] = 4
monthly_meteo_df.loc[idx_aut3, "m"] = 4 

## group by season and average except for precipitation 

monthly_meteo_df_noP = monthly_meteo_df.drop(["RR","dm"],axis = 1)
monthly_meteo_df_noP = monthly_meteo_df_noP.groupby(["m"]).mean()

monthly_meteo_df_P = monthly_meteo_df.drop(["G_Dh","dm", "G_h", "Ta", "Ts"],axis = 1)
monthly_meteo_df_P = monthly_meteo_df_P.groupby(["m"]).sum()

monthly_meteo_df_P

monthly_meteo_final = pd.merge(monthly_meteo_df_noP, monthly_meteo_df_P, how ="left", left_index = True, right_index = True)
monthly_meteo_final
```

# Weekday

```python

weekday_meteo_df = meteo_df.drop(["h"], axis = 1)

# weekday : monday, tuesday, wednesday, thursday, friday (0,1,2,3,4)
#2019 starts and ends on tuesday. For another year, one must find the first week day of the year.
length_dataframe = len(weekday_meteo_df)

idx_tue = np.arange(1,length_dataframe,168).tolist()
idx_wed = np.arange(25,length_dataframe,168).tolist()
idx_thu = np.arange(49,length_dataframe,168).tolist()
idx_fri = np.arange(73,length_dataframe,168).tolist()
idx_sat = np.arange(97,length_dataframe,168).tolist()
idx_sun = np.arange(121,length_dataframe,168).tolist()
idx_mon = np.arange(145,length_dataframe,168).tolist()

for i in range(len(idx_tue)):
    weekday_meteo_df.iloc[idx_tue[i]:idx_tue[i]+24, 0] = 1
    
for i in range(len(idx_wed)):
    weekday_meteo_df.iloc[idx_wed[i]:idx_wed[i]+24, 0] = 1
    weekday_meteo_df.iloc[idx_thu[i]:idx_thu[i]+24, 0] = 1
    weekday_meteo_df.iloc[idx_fri[i]:idx_fri[i]+24, 0] = 1
    weekday_meteo_df.iloc[idx_sat[i]:idx_sat[i]+24, 0] = 2
    weekday_meteo_df.iloc[idx_sun[i]:idx_sun[i]+24, 0] = 2
    weekday_meteo_df.iloc[idx_mon[i]:idx_mon[i]+24, 0] = 1
    
## group by weekday/weekend and average except for precipitation 

weekday_meteo_df_noP = weekday_meteo_df.drop(["RR","m"],axis = 1)
weekday_meteo_df_noP = weekday_meteo_df_noP.groupby(["dm"]).mean()

weekday_meteo_df_P = weekday_meteo_df.drop(["G_Dh", "G_h", "Ta", "Ts","m"],axis = 1)
weekday_meteo_df_P = weekday_meteo_df_P.groupby(["dm"]).sum()

weekday_meteo_final = pd.merge(weekday_meteo_df_noP, weekday_meteo_df_P, how ="left", left_index = True, right_index = True)
weekday_meteo_final    


```

# Day vs Night

```python
# Add seasons

season_meteo_df = meteo_df.drop(["dm"], axis=1)
daytime_meteo_df = meteo_df.drop(["dm"], axis=1)

# winter season : december, january, february (12,1,2)

idx_win = meteo_df[meteo_df["m"] == 12].index
idx_win2 = meteo_df[meteo_df["m"] == 1].index
idx_win3 = meteo_df[meteo_df["m"] == 2].index

season_meteo_df.loc[idx_win, "m"] = 1
season_meteo_df.loc[idx_win2, "m"] = 1
season_meteo_df.loc[idx_win3, "m"] = 1


# spring season : march, april, may
idx_spr = meteo_df[meteo_df["m"] == 3].index
idx_spr2 = meteo_df[meteo_df["m"] == 4].index
idx_spr3 = meteo_df[meteo_df["m"] == 5].index

season_meteo_df.loc[idx_spr, "m"] = 2
season_meteo_df.loc[idx_spr2, "m"] = 2
season_meteo_df.loc[idx_spr3, "m"] = 2

# summer season : june, july, august
idx_sum = meteo_df[meteo_df["m"] == 6].index
idx_sum2 = meteo_df[meteo_df["m"] == 7].index
idx_sum3 = meteo_df[meteo_df["m"] == 8].index

season_meteo_df.loc[idx_sum, "m"] = 3
season_meteo_df.loc[idx_sum2, "m"] = 3
season_meteo_df.loc[idx_sum3, "m"] = 3

# FAAALL season : september, october, november
idx_aut = meteo_df[meteo_df["m"] == 9].index
idx_aut2 = meteo_df[meteo_df["m"] == 10].index
idx_aut3 = meteo_df[meteo_df["m"] == 11].index

season_meteo_df.loc[idx_aut, "m"] = 4
season_meteo_df.loc[idx_aut2, "m"] = 4
season_meteo_df.loc[idx_aut3, "m"] = 4 

season_meteo_df
```

```python
# Add daytime depending on season

# Winter (1): day from 8am to 5pm --> 1 for day, 2 for night
idx_win = season_meteo_df[season_meteo_df["m"] == 1].index
winter_values = season_meteo_df.iloc[idx_win, :]
idx_win_night1 = winter_values[winter_values["h"]<=8].index
idx_win_night2 = winter_values[winter_values["h"]>=18].index

daytime_meteo_df.loc[idx_win_night1, "h"] = 2
daytime_meteo_df.loc[idx_win_night2, "h"] = 2

# Spring (2): day from 7am to 7pm
idx_spr = season_meteo_df[season_meteo_df["m"] == 2].index
spring_values = season_meteo_df.iloc[idx_spr, :]
idx_spr_night1 = spring_values[spring_values["h"]<=7].index
idx_spr_night2 = spring_values[spring_values["h"]>=20].index

daytime_meteo_df.loc[idx_spr_night1, "h"] = 2
daytime_meteo_df.loc[idx_spr_night2, "h"] = 2

# Summer (3): day from 6am to 9pm
idx_sum = season_meteo_df[season_meteo_df["m"] ==3].index
summer_values = season_meteo_df.iloc[idx_sum ,:]
idx_sum_night1 = summer_values[summer_values["h"]<=6].index
idx_sum_night2 = summer_values[summer_values["h"]>=22].index

daytime_meteo_df.loc[idx_sum_night1, "h"] = 2
daytime_meteo_df.loc[idx_sum_night2, "h"] = 2

# Autumn (4): day from 7am to 7pm
idx_aut = season_meteo_df[season_meteo_df["m"] ==4].index
aut_values = season_meteo_df.iloc[idx_aut, :]
idx_aut_night1 = aut_values[aut_values["h"]<=7].index
idx_aut_night2 = aut_values[aut_values["h"]>=20].index

daytime_meteo_df.loc[idx_aut_night1, "h"] = 2
daytime_meteo_df.loc[idx_aut_night2, "h"] = 2


#Set all others to day --> as 1
idx_day = daytime_meteo_df[daytime_meteo_df["h"] !=2].index
daytime_meteo_df.loc[idx_day, "h"] = 1

daytime_meteo_df
```

```python
# For all except precipitation, calculate avg by daytime and season
daytime_meteo_df_noP = daytime_meteo_df.drop(["RR"],axis = 1)
daytime_meteo_df_noP = daytime_meteo_df_noP.groupby(["h"]).mean()

# For precipitation, calculate sum by daytime and season
daytime_meteo_df_P = daytime_meteo_df.drop(["G_Dh", "G_h", "Ta", "Ts"],axis = 1)
daytime_meteo_df_P = daytime_meteo_df_P.groupby(["h"]).sum()

daytime_meteo_final = pd.merge(daytime_meteo_df_noP, daytime_meteo_df_P, how ="left", left_index = True, right_index = True)
daytime_meteo_final
```
